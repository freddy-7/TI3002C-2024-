```python
#Eliminar primera fila o filas innecesarias
df= pd.read_excel('Inmuebles.xlsx', skiprows=1)
df
```

```python
#Eliminar columnas innecesarias
#axis = 1 para columnas
#axis = 0 para filas
df= df.drop(['Referencia'], axis=1)
df
```

```python
#Conversión de tipo de dato de columna de tipo Object a Float
df['FECHA_ENT']= df['FECHA_ENT'].astype(str)
```

```python
#Reemplazo de datos tipo string a datos tipo numérico
df.CVE_CLPV = df.CVE_CLPV.replace({'M':'0'}, regex=False)
```

```python
#Analizar categorias de una columna
CATS_FECHA_ENT = df.groupby(['CVE_CLPV'])['CVE_CLPV'].count()
CATS_FECHA_ENT
```

```python
#Filtro por objeto
filtro1=df[(df["FECHA_ENT"] == '2019-09-03 00:00:00')]
filtro1
```
